```python
from pyspark.sql.functions import avg, max, min, count

silver_df = spark.table("silver_weather")
```

```python
from pyspark.sql.functions import avg, max, min, count, round

gold_df = silver_df.groupBy(
    "city",
    "country",
    "weather"
).agg(
    round(avg("temperature"), 2).alias("avg_temperature"),
    round(avg("humidity"), 2).alias("avg_humidity"),
    round(avg("pressure"), 2).alias("avg_pressure"),
    round(avg("wind_speed"), 2).alias("avg_wind_speed"),
    round(max("temperature"), 2).alias("max_temperature"),
    round(min("temperature"), 2).alias("min_temperature"),
    count("*").alias("total_records")
)

display(gold_df)
```

```python
gold_df.write \
    .format("delta") \
    .mode("overwrite") \
    .save("s3a://weather-pipeline-036815016562-ap-south-1-an/gold/weather_summary")
```

```python
%sql
CREATE TABLE IF NOT EXISTS gold_weather_summary
USING DELTA
LOCATION 's3a://weather-pipeline-036815016562-ap-south-1-an/gold/weather_summary';
```
